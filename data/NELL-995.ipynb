```python
import numpy
import csv
from collections import defaultdict as ddict
csv_file_path = "./NELL-995_onto/Schema-NELL.csv"
```

```python

with open(csv_file_path, 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter='\t')
    ontotriples = []
    # 跳过第一行（标题行）
    next(reader)
    
    for row in reader:
        if len(row) >= 3:  # 确保行中至少有3个字段
            subject = row[0].split(':')[1]
            predicate = row[1].split(':')[1]
            
            # 打印行内容以进行调试
            #print("Row:", row)
            
            object_parts = row[2].split(':')
            if len(object_parts) >= 2:
                object = object_parts[1]
                
                triple = (subject, predicate, object)
                ontotriples.append(triple)
            else:
                print("行数据不需要:", row)  # 打印不完整的行数据以进行调试

# 将三元组保存到文件
output_file_path = './NELL-995_onto/ontotriples.txt'  # 指定保存路径和文件名
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    for triple in ontotriples:
        output_file.write('\t'.join(triple) + '\n')

print("三元组已保存到", output_file_path)






```

```python
with open('./NELL-995_onto/ontotriples.txt', 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter='\t')
    subclass_triples = []

    for row in reader:
        subject = row[0]
        predicate = row[1]
        if predicate == 'subClassOf':  # 只提取关系为 "subClassOf" 的三元组
            object = row[2]
            triple = (subject, predicate, object)
            subclass_triples.append(triple)

with open('./NELL-995_onto/hierarchy.txt', 'w', encoding='utf-8') as hierarchy_file:
    for line in subclass_triples:
        hierarchy_file.write('\t'.join(line) + '\n')
        
```

```python

all_entity_ontologies,relations,ontologies,entities = [],[],[],[]
miss = []
for i,cleaned_line in enumerate(all_data):
    head, relation, tail =cleaned_line.strip().split('\t')# 去除末尾的换行符
    #print(head,relation,tail)
    #print(head)
    if len(head.split(':')) == 3:
        _, ontology1, entity1 = head.split(':')
        entities.append(ontology1+":"+entity1)
        #print(ontology1, entity1)
        all_entity_ontologies.append((entity1,ontology1))
        ontologies.append(ontology1)
    else:
        miss.append(head)
    

    if len(tail.split(':')) == 3:
        _, ontology2, entity2 = tail.split(':')
        all_entity_ontologies.append((entity2,ontology2))
        entities.append(ontology2+":"+entity2)
        ontologies.append(ontology2)
    else:
        miss.append(tail)
    relations.append(relation)

miss = set(miss)    
entities.extend(miss)
entities = list(set(entities))
ontologies.append("data_propetity")

relations = list(set(relations))
ontologies = list(set(ontologies))
for i in miss:
    all_entity_ontologies.append("data_propetity:"+i)
all_entity_ontologies=list(set(all_entity_ontologies))

print("nums of relations:",len(relations))
print("nums of ontologies:",len(ontologies))
print("nums of entities:",len(entities))
print("nums of all_entity_ontologies",len(all_entity_ontologies))
print("nums of data_propeties:",len(miss))

```

```python
ont_reidx = dict()
ontidx = 0
for ontology in ontologies:
    ont_reidx[ontology] = ontidx
    ontidx += 1 

with open('./ont2id', 'w') as file:
    for key, value in ont_reidx.items():
        file.write(f"{key}\t{value}\n")
    
```

```python
def reidx(triples):
    ent_reidx = dict()
    rel_reidx = dict()
    
    entidx = 0
    relidx = 0
    
    reidx_triples = []
    for tri in triples:

        h, r, t = tri.strip().split("\t")
        if h not in ent_reidx.keys():
            ent_reidx[h] = entidx
            entidx += 1
        if t not in ent_reidx.keys():
            ent_reidx[t] = entidx
            entidx += 1
        if r not in rel_reidx.keys():
            rel_reidx[r] = relidx
            relidx += 1
    
        reidx_triples.append((ent_reidx[h], rel_reidx[r], ent_reidx[t]))
    
    return reidx_triples, ent_reidx, rel_reidx
    
```

```python

triples, ent_reidx, rel_reidx = reidx(all_data)
ent_reidx_inv = {v: k for k, v in ent_reidx.items()}
rel_reidx_inv = {v: k for k, v in rel_reidx.items()}
with open('./en2id', 'w') as file:
    for key, value in ent_reidx.items():
        file.write(f"{key}\t{value}\n")

with open('./rel2id', 'w') as file:
    for key, value in rel_reidx.items():
        file.write(f"{key}\t{value}\n")
```

```python
ent2ont = {}
ont2ent = ddict(list)
for ents, idx in ent_reidx.items():
    #print(ents)
    if len(ents.split(":")) == 3:
        ont, ent = ents.split(":")[1:]
        ent2ont[idx] = ont_reidx[ont]
    else:
        ent2ont[idx] = ont_reidx['data_propetity']
    
for ont,idx in ont_reidx.items():
    ont2ent[idx]=[key for key, value in ent2ont.items() if value == idx]

```
